```python
#order total vs cuisine
plt.figure(figsize=(15,5))
sns.boxplot(data=df, x='cuisine_type', y='cost_of_the_order')
plt.xticks(rotation=60)
plt.show()
```

```python
#delivery time vs day
plt.figure(figsize=(15,5))
sns.boxplot(data=df, x='day_of_the_week', y='delivery_time')
plt.xticks(rotation=60)
plt.show()
```

```python
# heatmap to show correlation among variables
vals = ['cost_of_the_order','delivery_time','food_preparation_time']
plt.figure(figsize=(15,5))
sns.heatmap(df[vals].corr(), annot=True)
plt.show()
```

```python
df_rating = df[df['rating'] != 'Not given'].copy() #showing only the restaurants with given ratings

#convert rating(obj) --> rating(int)
df_rating['rating'] = df_rating['rating'].astype(int)

#restaurants with their rating count
df_rating_rest = df_rating.groupby(['restaurant_name'])['rating'].count().sort_values(ascending=False).reset_index()
df_rating_rest.head()
```

```python
df['total time'] = df['food_preparation_time'] + df['delivery_time']

# orders that take more than 60 mins to be delivered
df[df['total time'] > 60] * 100
```
